```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Final_Housing_Train_Data.csv")
```

```python
df.head()
```

```python
df['SalePrice'].mean()
```

```python
train_df = pd.read_csv("train.csv")
```

```python
# sns.pairplot(data=train_df,hue='SalePrice')
```

```python
X = df.drop('SalePrice',axis=1)
```

```python
X
```

```python
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# tring predictions with ADA Booster model
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import accuracy_score,classification_report
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
model = AdaBoostRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_prediction = model.predict(X_test)
```

```python
mean_absolute_error(y_test,y_prediction)
```

```python
np.sqrt(mean_squared_error(y_test,y_prediction))
```

```python
# Using Random Forest for predictions
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
n_estimators = [10,50,100,200]
bootstrap = [True,False]
oob_score = [True,False]

grid_param = {'n_estimators':n_estimators,'bootstrap':bootstrap,'oob_score':oob_score}
```

```python
model = RandomForestRegressor()
```

```python
grid_model = GridSearchCV(estimator=model,param_grid=grid_param)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
y_predictions = grid_model.predict(X_test)
```

```python
mean_absolute_error(y_test,y_predictions)
```

```python
model = RandomForestRegressor(n_estimators=200)
```

```python
model.fit(X_train,y_train)
```

```python
y_predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test,y_predictions)
```

```python
np.sqrt(mean_squared_error(y_test,y_predictions))
```
